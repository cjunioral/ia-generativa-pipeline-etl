```python
from openai import OpenAI

client = OpenAI()

def generate_ai_news(product):
    
    try:
        prompt = f'''
    Você é um especialista em marketing de e-commerce.
    Crie uma frase curta em português, com no máximo 120 caracteres,
    sobre este produto, sem inventar informações.

    Produto: {product['title']}
    Categoria: {product['category']}
    Preço: {product['price']}
    Desconto: {product['discountPercentage']}%
    Avaliação: {product['rating']}/5
    '''.strip()

        response = client.responses.create(
            model='gpt-4o-mini',
            input=prompt
        )
        
        return response.output_text.strip()
        
    except Exception:
        print('Falha ao usar o OpenAI')
```

```python
sample = df.iloc[0].to_dict()
generate_ai_news(sample)
```
